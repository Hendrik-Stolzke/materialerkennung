```python
import pandas as pd    
from pandas import DataFrame            
import glob
from tsfresh.feature_extraction import extract_features, MinimalFCParameters
from tsfresh.feature_selection import select_features
```

```python
# list all csv files only
csv_files = glob.glob('*.{}'.format('csv'))
csv_files = csv_files[0:-1]
```

```python
measure = pd.read_csv('measurements.csv')
df_measure = DataFrame(measure)
df_measureMat = df_measure.material
df_measures = DataFrame(df_measureMat)
df_measures = df_measures.material.tolist()
measures_Series = pd.Series(df_measures)
```

```python
l = [] 
i = 0
for file in csv_files:
    target = df_measures[i]
    data = pd.read_csv(file)
    df = DataFrame(data)
    df['id'] = i
    df['target'] = target
    l.append(df)
    i += 1
```

```python
df_full = DataFrame(columns=['Audio', 'Voltage', 'Current','id'])
df_full
for elem in l:
    df_full = pd.concat([df_full, elem], ignore_index=True)
```

```python
df2 = df_full.drop(columns=['target'])
```

```python
extraction_settings = MinimalFCParameters()
extracted_features = extract_features(df2, column_id = 'id', default_fc_parameters=extraction_settings)
extracted_features.info()
```

```python
X_selected = select_features(extracted_features, measures_Series)
```

```python
X_selected.drop(['Voltage__mean','Voltage__root_mean_square', 'Current__root_mean_square', 'Current__mean', 'Audio__variance'], axis = 1,  inplace=True)
X_selected
```
